```python
import numpy as np
import matplotlib.pyplot as plt

steps=200
t=np.arange(0,steps)
x=np.zeros(steps)
y=np.zeros(steps)

for i in range(1,steps):
    ran=np.random.choice(['up','down','left','right'])
    if ran == 'up':
        x[i]=x[i-1]
        y[i]=y[i-1]+1
    elif ran == 'down':
        x[i]=x[i-1]
        y[i]=y[i-1]-1
    elif ran == 'left':
        x[i]=x[i-1]-1
        y[i]=y[i-1]
    elif ran == 'right':
        x[i]=x[i-1]+1
        y[i]=y[i-1]

#cumulative sume of x and y
cumx=np.cumsum(x)
cumy=np.cumsum(y)


plt.plot(x,y,'g*-')
plt.xlabel('x')
plt.ylabel('y')
plt.grid('true')
plt.title('2D random walk')
plt.axis('equal')
plt.savefig('randomwalk.png')
```
